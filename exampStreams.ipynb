```python
import websocket
import json
import ssl

def on_error(ws, error):
    print(error)

def on_close(ws):
    print("### closed ###")
    #connection.close()

def on_open(ws):
    print("ONOPEN")
    ws.send(json.dumps({'command':'subscribe','channel':'BTC_ETH'}))

def on_message(ws, message):
    message = json.loads(message)
    print(message)
                
websocket.enableTrace(True)
context1 = ssl._create_unverified_context()
ws = websocket.WebSocketApp("wss://api2.poloniex.com/",
                              on_message = on_message,
                              on_error = on_error,
                              on_close = on_close
                              )
ws.on_open = on_open
ws.run_forever()
```
